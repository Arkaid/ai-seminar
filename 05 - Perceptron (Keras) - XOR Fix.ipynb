# パーセプトロン（Keras版）

先の手作りのパーセプトロンをニューラルネットワーク専用のツール「Keras」で構築し、どれぐらい簡単になるのか調べてみよう

まず、必要なパッケージを追加しよう

```python
import numpy as np

# 入力、結合、活性化の「レイヤ」
from keras.layers import Input, Dense, Activation

# モデル
from keras.models import Model

# 活性化関数
from keras.activations import sigmoid
```

そうしたら、Kerasでパーセプトロンを作成

```python
def create_perceptron():
    
    # 入力レイヤ、shapeは(2,) - 入力が２つ、サンプル数が今決まらない
    x = Input(shape=(2,), dtype=float)
    
    # 結合レイヤ（荷重）出力が１つ
    w = Dense(units=1)(x)
    
    # 活性化レイヤ。活性化関数はsigmoid
    y = Activation(sigmoid)(w)
    
    # モデル
    model = Model(inputs=x, outputs=y)
    
    return model
```

```python
# 確認しましょう
model = create_perceptron()
model.summary()
```

良くみると、`Trainable params`（学習できる数値）は３つがある。これは荷重２つとバイアスである

```python
# 実験を統一するため、もう一度乱数のシードを固定
np.random.seed(12345)

# モデルを再構築し、中身を見よう
model = create_perceptron()
print (model.get_weights())
```

このまま（学習なし）で推定をしてみると：

```python
x = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
], dtype=float)

y_pred = model.predict(x)
print(y_pred)
```

まったく当たらないので、学習させましょう！

まず、kerasでは、そのまま学習できない。その先、`create_perceptron()`で作ったモデルを「コンパイル」しなければならない。コンパイルすることにて、モデルをGPUで実行できるようになる

```python
# ここで損失関数を設定：mean_squared_errorは平均二乗誤差
# 「sgd」は「Stochastic Gradient Descent」の意味で、optimizer（最適化方法）については後回しにしましょう～
model.compile(optimizer="sgd", loss="mean_squared_error")
```

```python
# 学習データを用意
y_true = np.array([ 0, 0, 0, 1 ], dtype=float)

# 学習実行
model.fit(x, y_true, epochs=5000)
```

```python
#結果を確認
y_pred = model.predict(x)
print(y_pred)
```

あれ？先よりも結果が悪いなぁ…

## 改善方法１：batchのサイズを選ぶ

先自分で作ったパーセプトロンを学習したときに、確かに、学習データ全て（４点）一発で学習したのが、誤差を計算したのは、1個ずつだった。
この学習方法で、小まめに荷重を調整できるが、学習時間がのびる。

kerasはデフォルトとして、学習データを全てを渡しても、誤差を1個ずつに計算せず、32点ずつのまとめた誤差を使う。比較的に処理が速いが、荷重の調整が大変

今回のAIは単純で、1個ずつで学習させましょう。これが「`fit`」をするときに`batch_size`を設定すればできる

```python
# 実験を統一するため、もう一度乱数のシードを固定
np.random.seed(12345)

# モデルを再構築し、中身を見よう
model = create_perceptron()
model.compile(optimizer="sgd", loss="mean_squared_error")
```

```python
# 学習実行
model.fit(x, y_true, batch_size=1, epochs=5000)
```

```python
#結果を確認
y_pred = model.predict(x)
print(y_pred)
```

よくなったが、まだスッキリしないなぁ

## 改善方法２：適する損失関数を選ぶ

「平均二乗誤差」普段に「距離」が表している（正解数値からどれぐらい離れるのか？）。実数の場合はこれで良いと思っていいのが、学習データはバイナリ（0か1）の場合、kerasがもと良い損失関数を提供する。それは`binary_crossentropy`という。

それでは、関数を変えて、確かめてみよう～

```python
# もう一度乱数を固定、モデルを構築する
np.random.seed(12345)
model = create_perceptron()
model.compile(optimizer="sgd", loss="binary_crossentropy")
```

```python
# 学習実行
model.fit(x, y_true, batch_size=1, epochs=5000)
```

```python
#結果を確認
y_pred = model.predict(x)
print(y_pred)
```

```python
print(model.get_weights())
```

## XORの修正：ネットワークを深くする

```python
# パーセプトロンではなくなった！
def create_model():
    
    x = Input(shape=(2,), dtype=float)
    h = Dense(units=2, activation=sigmoid)(x)
    y = Dense(units=1, activation=sigmoid)(h)
    
    model = Model(inputs=x, outputs=y)

    return model
```

```python
# もう一度乱数を固定、モデルを構築する
model = create_model()
model.compile(optimizer="sgd", loss="binary_crossentropy")

model.summary()
```

```python
# 学習データを用意
y_true = np.array([ 0, 1, 1, 0 ], dtype=float)

# 学習実行
model.fit(x, y_true, batch_size=1, epochs=5000)
```

```python
#結果を確認
y_pred = model.predict(x)
print(y_pred)
```

## 中間層の問題：活性化

入力層の数値は0～1にしても。中間層だとそれに限らない
sigmoid活性化の出力はいつも0～１になるので、中間層があまりうまく学習できない。その変わり、-1 ～ 1の出力ができるtanh活性化を使おう！

### sigmoid
<img src="./sigmoid.png" width=650/>

### tanh
<img src="./tanh.png" width=650/>

```python
# tanhを変えてみよう
from keras.activations import tanh

def create_model():
    
    x = Input(shape=(2,), dtype=float)
    h = Dense(units=2, activation=tanh)(x)
    y = Dense(units=1, activation=sigmoid)(h)
    
    model = Model(inputs=x, outputs=y)

    return model
```

```python
# もう一度乱数を固定、モデルを構築する
model = create_model()
model.compile(optimizer="sgd", loss="binary_crossentropy")
```

```python
# 学習実行
model.fit(x, y_true, batch_size=1, epochs=5000)
```

```python
#結果を確認
y_pred = model.predict(x)
print(y_pred)
```
