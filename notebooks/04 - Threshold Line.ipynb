# 荷重とバイアスで構築できる直線

入力が２つだけのパーセプトロンがパラメータ３つがある。
このパラメータで2次元空間でも直線を作成できる。この線がデータを２群に分割する（分類）

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 入力データの可視化

```python
# ２次元の図で表す
x = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
], dtype=float)

plt.figure(figsize=(6,6))
plt.scatter(x[:,0], x[:,1], c=['red', 'green', 'blue', 'black'])
```

```python
# 直線上に２つの点を調べる
def calculate_line(w1, w2, b):
    x1 = -b / w1 # x2 = 0
    x2 = -b / w2 # x1 = 0
    
    pt1 = [x1, 0]
    pt2 = [0, x2]
    
    return pt1, pt2
    
# 4点と直線を表示 
def plot_line(pt1, pt2, colors):
    plt.figure(figsize=(6,6))
    plt.scatter(x[:,0], x[:,1], c=colors)
    plt.plot(pt1, pt2)
```

## ANDの場合

```python
# and の荷重とバイアスを試す：
w1 = 4.234712
w2 = 4.2333617
b = -6.542434

pt1, pt2 = calculate_line(w1, w2, b)
print(pt1)
print(pt2)
```

```python
plot_line(pt1, pt2, ['red', 'red', 'red', 'green'])
```

## ORの場合

```python
# and の荷重とバイアスを試す：
w1 = 5.365714
w2 = 5.3412714
b = -2.1799393

pt1, pt2 = calculate_line(w1, w2, b)
print(pt1)
print(pt2)
```

```python
plot_line(pt1, pt2, ['red', 'green', 'green', 'green'])
```
